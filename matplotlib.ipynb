**subplot**

```python
import matplotlib.pyplot as plt
import numpy as np
#plot1
x = np.array([1, 2, 3, 4,5])
y = np.array([1, 2, 3, 4,5])
font1 = {'family':'serif','color':'blue','size':10}
font2 = {'family':'serif','color': 'red','size':15}
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title('first graph',fontdict=font1)
#plot2
x = np.array([1, 2, 3, 4,5])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title('second graph',fontdict=font1)
#plot3
x = np.array([1, 2, 3, 4,5])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title('third graph',fontdict=font1)
#plot4
x = np.array([5, 4, 3, 2,1])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title('fourth graph',fontdict=font1)
#subplot Adjust
plt.subplots_adjust(wspace=0.4,hspace=0.4)
#supertittle
plt.suptitle('all graphs',fontdict=font2)
plt.show()
```

**Matplotlib Pie Charts**

**Creating Pie Charts**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

**Labels**

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show()
```

**Start Angle**

```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

**Explode**

```python
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

**Shadow**

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

COLORS

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

**LEGEND**

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels,colors=mycolors)
plt.legend()
plt.show()
```

**Legend With Header**

```python
plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
