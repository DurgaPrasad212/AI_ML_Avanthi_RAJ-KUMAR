```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

LINE CHART:

```python
x=[1,2,3]
y=[4,5,6]

plt.figure(figsize=(5,3),dpi=100)
plt.title('Our first sine Graph',fontdict={'fontsize':10})
plt.plot(x,y)
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()

```

```python
x=np.arange(0,5,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('our first sin graph',fontdict ={'fontsize':35})
```

Q1. EXAMPLE

```python
x = [2,4,6,8,10,12]
y = [3,6,9,12,15,18]
plt.plot(x,y)
plt.title('my first graph',fontdict ={'fontsize':10})
plt.xlabel('x cordinates')
plt.ylabel('y cordinates')
plt.xticks([2,4,6,8,10,12])
plt.yticks([3,6,9,12,15,18])
plt.legend()
```

```python
x=[1,2,3]
y=[4,5,6]
plt.figure(figsize=(5,3),dpi=100)
plt.plot(x,y)
plt.title('our first  graph',fontdict ={'fontsize':20})
plt.xlabel('x cordinates')
plt.ylabel('y cordinates')
x2=np.arange(0,5,1)
plt.plot(x2,x2**2,'y',label='x^2')

```
