```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/canada_per_capita_income (3).csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='o')
plt.show()
```

```python
X = df.drop('per capita income (US$)', axis = 'columns')
X
```

```python
y = df['per capita income (US$)']
y
```

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
model2.fit(X,y)
```

```python
model2.predict([[2024]])
```

```python
model2.score(X,y)
```

```python
model2.coef_
```

```python
model2.intercept_
```
